```python
def main():
    introduction = intro()
    board = create_grid()
    pretty = printPretty(board)
    symbol_1, symbol_2 = sym()
    full = isFull(board, symbol_1, symbol_2)




def intro():
    print("Vítejte ve hře piškvorky!")
    print("\n")
    print("Pravidla: Hráč 1 a hráč 2, reprezentovaní symboly X a O, se střídají v označování políček "
          "v mřížce 3x3. Hráč, který jako první umístí tři své symboly do řady vodorovně, svisle nebo diagonálně, vyhrává.")
    print("\n")
    input("Stiskněte Enter pro pokračování.")
    print("\n")



def create_grid():
    # herní plocha
    print("Herní plocha:")
    board = [[" ", " ", " "],
             [" ", " ", " "],
             [" ", " ", " "]]
    return board


def sym():
    # výběr symbolu
    while True:
        symbol_1 = input("Hráči 1, chcete být X nebo O? ").upper()
        if symbol_1 in ["X", "O"]:
            if symbol_1 == "X":
                symbol_2 = "O"
            else:
                symbol_2 = "X"
            print(f"Hráči 1, jste {symbol_1}.")
            print(f"Hráči 2, jste {symbol_2}.")
            input("Stiskněte Enter pro pokračování.")
            print("\n")
            return symbol_1, symbol_2
        else:
            print("Neplatný vstup. Vyberte buď 'X', nebo 'O'.")


def startGamming(board, symbol_1, symbol_2, count):
    # začátek hry

    # kdo je na tahu
    if count % 2 == 0:
        player = symbol_1
    else:
        player = symbol_2
    print(f"Hráči {player}, jste na tahu.")
    row = int(input("Vyberte řádek (0 - horní, 1 - střední, 2 - dolní): "))
    column = int(input("Vyberte sloupec (0 - levý, 1 - střední, 2 - pravý): "))

    # je výběr v rozsahu
    while (row > 2 or row < 0) or (column > 2 or column < 0):
        outOfBoard(row, column)
        row = int(input("Vyberte řádek (0 - horní, 1 - střední, 2 - dolní): "))
        column = int(input("Vyberte sloupec (0 - levý, 1 - střední, 2 - pravý): "))

    # je políčko už obsazené
    while board[row][column] in [symbol_1, symbol_2]:
        illegal(board, symbol_1, symbol_2, row, column)
        row = int(input("Vyberte řádek (0 - horní, 1 - střední, 2 - dolní): "))
        column = int(input("Vyberte sloupec (0 - levý, 1 - střední, 2 - pravý): "))

    # umístění symbolu hráče
    board[row][column] = player
    return board


def isFull(board, symbol_1, symbol_2):
    count = 1
    winner = True
    # je plocha plná
    while count < 10 and winner:
        gaming = startGamming(board, symbol_1, symbol_2, count)
        pretty = printPretty(board)

        if count == 9:
            print("Herní plocha je plná. Konec hry.")
            if winner:
                print("Hra skončila remízou.")

        # vítěz
        winner = isWinner(board, symbol_1, symbol_2, count)
        count += 1

    if not winner:
        print("Konec hry.")

    # závěrečná zpráva
    report(count, winner, symbol_1, symbol_2)


def outOfBoard(row, column):
    print("Výběr je mimo rozsah. Zvolte jiné políčko.")


def printPretty(board):
    # výpis herní plochy
    rows = len(board)
    print("---+---+---")
    for r in range(rows):
        print(f"{board[r][0]} | {board[r][1]} | {board[r][2]}")
        print("---+---+---")
    return board


def isWinner(board, symbol_1, symbol_2, count):
    # vyhrál někdo
    winner = True

    # kontrola řádků
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] == symbol_1:
            winner = False
            print(f"Hráči {symbol_1}, vyhráváte!")
        elif board[row][0] == board[row][1] == board[row][2] == symbol_2:
            winner = False
            print(f"Hráči {symbol_2}, vyhráváte!")

    # kontrola sloupců
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] == symbol_1:
            winner = False
            print(f"Hráči {symbol_1}, vyhráváte!")
        elif board[0][col] == board[1][col] == board[2][col] == symbol_2:
            winner = False
            print(f"Hráči {symbol_2}, vyhráváte!")

    # kontrola diagonálně
    if board[0][0] == board[1][1] == board[2][2] == symbol_1:
        winner = False
        print(f"Hráči {symbol_1}, vyhráváte!")
    elif board[0][0] == board[1][1] == board[2][2] == symbol_2:
        winner = False
        print(f"Hráči {symbol_2}, vyhráváte!")
    elif board[0][2] == board[1][1] == board[2][0] == symbol_1:
        winner = False
        print(f"Hráči {symbol_1}, vyhráváte!")
    elif board[0][2] == board[1][1] == board[2][0] == symbol_2:
        winner = False
        print(f"Hráči {symbol_2}, vyhráváte!")

    return winner


def illegal(board, symbol_1, symbol_2, row, column):
    # informuje hráče že vybrané políčko je již obsazené
    print("Vybrané políčko je již obsazené. Zvolte jiné.")


def report(count, winner, symbol_1, symbol_2):
    # závěrečná zpráva o hře
    print("\n")
    input("Stiskněte Enter pro zobrazení shrnutí hry.")
    if not winner and count % 2 == 1:
        print(f"Vítěz: Hráč {symbol_1}.")
    elif not winner and count % 2 == 0:
        print(f"Vítěz: Hráč {symbol_2}.")
    else:
        print("Hra skončila remízou.")


main()
```
